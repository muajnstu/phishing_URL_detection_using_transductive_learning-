## Removing outlier from individual column


```python
# removing outliers using z-score methods
upper_limit = data['URLLength'].mean() + 3*data['URLLength'].std()
lower_limit = data['URLLength'].mean() - 3*data['URLLength'].std()
```

```python
data.loc[(data['URLLength'] >upper_limit) | (data['URLLength'] <lower_limit)  ]
```

```python
#trimming the outliers

new_data=data.loc[(data['URLLength'] <upper_limit) & (data['URLLength'] >lower_limit)  ]
```

```python
print('before removing outliers: ', len(data))
print('new data: ', len(new_data))
print('outliers: ', len(data)- len(new_data))
```

```python
## Removing outlier from multiple columns
```

```python

# Function to detect and remove outliers
def remove_outliers(df, columns, threshold=3):
    z_scores = np.abs((df[columns] - df[columns].mean()) / df[columns].std())
    mask = (z_scores < threshold).all(axis=1)
    return df[mask]

# Specify columns to detect outliers
columns_to_check = ['URLLength','DomainLength']

# Remove outliers
data_no_outliers = remove_outliers(data, columns_to_check)

# Display the result
print("Original DataFrame shape:", data.shape)
print("DataFrame shape after removing outliers:", data_no_outliers.shape)
```

## Removing outlier from entire dataset

```python
# Function to detect and remove outliers
def remove_outliers(df, threshold=3):
    z_scores = np.abs((df - df.mean()) / df.std())
    mask = (z_scores < threshold).all(axis=1)
    return df[mask]

# Remove outliers
data_no_outliers = remove_outliers(data)

# Display the result
print("Original DataFrame shape:", data.shape)
print("DataFrame shape after removing outliers:", data_no_outliers.shape)
```

```python


# Function to detect and remove outliers using IQR method
def remove_outliers_iqr(df, k=1.5):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - k * IQR
    upper_bound = Q3 + k * IQR
    return df[~((df < lower_bound) | (df > upper_bound)).any(axis=1)]

# Remove outliers
data_no_outliers = remove_outliers_iqr(data)

# Display the result
print("Original DataFrame shape:", data.shape)
print("DataFrame shape after removing outliers using IQR method:", data_no_outliers.shape)
```
